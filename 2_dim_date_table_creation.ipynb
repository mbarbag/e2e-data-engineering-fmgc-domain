```python
%sql
CREATE OR REPLACE TABLE fmcg.gold.dim_date AS
SELECT
  d AS month_start_date,
  TO_CHAR(d,'yyyyMM') AS date_key,
  YEAR(d) AS year,
  TO_CHAR(d, 'MMMM') AS month_name,
  TO_CHAR(d, 'MMM') AS month_short_name,
  concat('Q',quarter(d)) AS quarter,
  concat(year,'-',quarter) AS year_quarter
FROM (
  SELECT explode(
    sequence(
      to_date('2024-01-01'),
      to_date('2025-12-01'),
      interval 1 month
    )
  ) AS d
);
```
